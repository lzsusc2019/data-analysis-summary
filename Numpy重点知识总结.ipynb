## 布尔值索引(复制)
* 副本深复制ndarray.copy()函数创建一个深层副本。 它是数组及其数据的完整副本，不与原始数组共享。

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.random.randn(7, 4)
```

```python
names
```

```python
data
```

```python
names == 'Bob' #会与一个Array里面的值一一比较
```

```python
#This boolean array can be passed when indexing the array

data[names == 'Bob'] #返回True对应的数组
#boolean索引的长度必须一致
#长度不匹配不会失败，但是要谨慎处理
```

```python
data[names == 'Bob', 2:]
```

```python
data[data < 0] = 0
data
```

```python
data[names != 'Joe'] =7
data
```

## 花式索引(复制)

```python
arr = np.empty((8, 4))

for i in range(8):
    arr[i] = i
arr
```

```python
arr[[4, 3, 0, 6]] #编号4，3，0，6的行，索引号
```

```python
arr = np.arange(32).reshape((8,4))
arr
```

```python
arr[[1, 5, 7, 2],[0, 3, 1, 2]] #抓编号(1,0) (5,3) ,(7,1),(2,2)
```

```python
arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]]#先选取索引号为1，5，7，2的行，然后再对行进行选值
```

# 维度与轴
* 转置，换轴

```python
arr = np.arange(15).reshape((3, 5))
arr
```

```python
arr.T
```

```python
np.dot(arr.T, arr) #(3, 6) * (6, 3) = (3, 3)
```

```python
arr = np.arange(16).reshape(2, 2, 4)
arr
```

```python
arr.transpose((1, 0, 2))
```

```python
arr.swapaxes(1, 2) #轴对调 (2,2,4) 变成 (2,4,2)
```

* reshape,ravel,flatten

```python
arr = np.arange(3*4).reshape((4,3), order='C') #行优先
arr
```

```python
brr = np.arange(3*4).reshape((4,3), order='F')#列优先
brr
```

```python
#扁平化
arr = np.arange(3*5).reshape(3, 5)
arr
```

```python
arr.ravel()[1] = 999 #扁平化操作
arr
```

```python
arr.ravel() #视图
```

```python
arr = np.arange(3*5).reshape(3, 5)
arr
arr.flatten() #也是扁平化，区别在于flatten 展开后是原来数据的copy，但是ravel 是原来数据的视图
```

# 广播
* 广播的原则：如果两个数组的后缘维度(从末尾开始算起的维度)的轴长度相符或者其中一方的长度为1。广播会在缺失和(或)长度为1的维度上进行

```python
c = np.arange(12).reshape(4, 3)
c
```

```python
d = np.random.rand(4, 3)
d
```

```python
e = np.random.randn(3, 3)
e
```

```python
c + d
```

```python
c + e #不符
```

```python
c + np.random.randn(3) #广播 (3,)
```

```python
c + np.random.randn(4) #不可以广播，不符合
```

```python
f = np.random.randn(2, 3, 4, 5)
```

```python
e = f + np.random.randn(2, 1, 1, 5) 
```
